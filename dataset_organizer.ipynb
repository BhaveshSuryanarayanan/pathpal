```python
import os
import shutil

# base_number = 469
base_number = 2182
# base_number = 1900
# base_number = 0

# Specify the directory containing your dataset
train_images_path = test_path  # Path to the folder containing .jpg files
train_labels_path = test_path  # Path to the folder containing .xml files
output_base_dir = r"/home/bhavesh/pathpal/datasets/grocery/dataset6_organized/test"

# Specify the desired prefix
prefix = "dataset6"
# Create output directories
output_image_dir = os.path.join(output_base_dir, "images")
output_annotation_dir = os.path.join(output_base_dir, "labels")
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_annotation_dir, exist_ok=True)

# Get sorted lists of files
image_files = sorted([f for f in os.listdir(train_images_path) if f.endswith('.jpg')])
annotation_files = sorted([f for f in os.listdir(train_labels_path) if f.endswith('.xml')])

# Check if the counts match
if len(image_files) != len(annotation_files):
    print("Error: The number of images and annotations does not match!")
    exit()

# Renaming and saving process
for i, (image, annotation) in enumerate(zip(image_files, annotation_files), start=1):
    new_image_name = f"{prefix}_image_{base_number + i : 08d}.jpg"
    new_annotation_name = f"{prefix}_image_{base_number + i: 08d}.xml"

    # Copy and rename image
    shutil.copy(
        os.path.join(train_images_path, image),
        os.path.join(output_image_dir, new_image_name)
    )

    # Copy and rename annotation
    shutil.copy(
        os.path.join(train_labels_path, annotation),
        os.path.join(output_annotation_dir, new_annotation_name)
    )

    print(f"Copied and renamed: {image} -> {new_image_name}, {annotation} -> {new_annotation_name}")

print(f"Renaming and saving complete! Files are in {output_base_dir}.")
```
