```python
import seaborn as sns
import matplotlib.pyplot as plt

def get_colors(num_classes):
    return sns.color_palette("hsv", num_classes)  # Try "Set3", "tab10", etc.

num_classes = 15
colors = get_colors(num_classes)

# Plot the colors
plt.figure(figsize=(num_classes, 1))
for i, color in enumerate(colors):
    plt.fill_between([i, i + 1], 0, 1, color=color)

plt.xlim(0, num_classes)
plt.yticks([])
plt.xticks(range(num_classes), [f'Class {i+1}' for i in range(num_classes)], rotation=45)
plt.title("Color Palette for Bounding Boxes")
plt.show()
```
